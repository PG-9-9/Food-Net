```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
import datetime
import os 

print(f"Notebook last run:{datetime.datetime.now()}")
print(f"Tensorflow version :{tf.__version__}")

# Downloading the helper functions 
if not os.path.exists("helper_functions.py"):
    !wget https://raw.githubusercontent.com/mrdbourke/tensorflow-deep-learning/main/extras/helper_functions.py
else:
    print("[INFO] helper_functions.py found in directory, skipping download")

from helper_functions import create_tensorboard_callback, plot_loss_curves, compare_historys
import tensorflow_datasets as tfds
```

```python
################ Loading the dataset
# All the available datasets
datasets_list=tfds.list_builders()

# Set our target datasets and see if it exists.
target_dataset="food101"
print(f"{target_dataset} in tensorflow datasets: {target_dataset in datasets_list}")

# Load the dataset
(train_data, test_data),ds_info=tfds.load(name="food101",
                                         split=["train", "validation"],
                                         shuffle_files=True,
                                         as_supervised=True,
                                        # True = tuple format 
                                        # False = Dictionary format,
                                          with_info=True
                                        # True = metadata -> tfds.load() returns tuple (data, ds_info)
                                        # False = metadata -> tfds.load() returns tuple (data)
                                         )
```

```python
# Exploring the data
print(ds_info.features)

class_names=ds_info.features["label"].names
print("********* First 10 class_names")
print(class_names[:10])

train_one_sample=train_data.take(1) #samples are in format(image_tensor, label)
print("*********Take one sample from the training_data")
print(train_one_sample)

print("*********Single element values")

for image, label in train_one_sample:
    print(f"""
    Image shape :{image.shape}
    Image dtype :{image.dtype}
    Target class from Food101 (tensor form):{label}
    Class form (str form): {class_names[label.numpy()]}
    """)

print("**********Checking the min & max values")
print(tf.reduce_min(image), tf.reduce_max(image))
# Values are 0 and 255 normalized.

print("******* Plotting the image")
import matplotlib.pyplot as plt
plt.imshow(image)
plt.title(class_names[label.numpy()])
plt.axis(False)

"""
    Steps to do before forming the dataloader:
    
    1. Need to convert uint8[downloaded type] -> float32
    2. Make all tensors of same size
    3. Scale values between 0 and 1 and Normalize the values of the images

"""
```


    # Steps to do before forming the dataloader:
    
    1. Need to convert uint8[downloaded type] -> float32
    2. Make all tensors of same size
    3. Scale values between 0 and 1 and Normalize the values of the images



```python
def preprocess_img(image, label, img_shape=224):
    """
        converts the image from uint8 into float32 
        and reshapes the image into [img_shape,img_shape, out_channels] 
    
    """
    # Making image of same size
    image=tf.image.resize(image, [img_shape,img_shape])
    # Convert into float
    return tf.cast(image,tf.float32),label
```

```python
preprocessed_img=preprocess_img(image,label)[0]
#plot the preprocessed image:
plt.imshow(preprocessed_img/255.)
plt.title(class_names[label])
plt.axis(False)
```

```python
# Mapping preprocessing function to training data and parallize
train_data=train_data.map(map_func=preprocess_img,
                         num_parallel_calls=tf.data.AUTOTUNE)

# Adding shuffle, batch , and prefetch to training data
train_data=train_data.shuffle(buffer_size=1000).batch(batch_size=32).prefetch(buffer_size=tf.data.AUTOTUNE)


# Adding map, batch, and prefetch to the test data,
# Shuffling is not required.
test_data=test_data.map(preprocess_img,num_parallel_calls=tf.data.AUTOTUNE)
test_data=test_data.batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
# Visualizing the test and train data
train_data, test_data
```

```python
# Creating callbacks:

# Tensorboard callbacks from the helper functions
from helper_functions import create_tensorboard_callback

# Modelcheckpoint callbacks
checkpoint_path="model_checkpoints/cp.ckpt"
model_checkpoint=tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                   monitor="val_accuracy",
                                                   save_best_only=True,
                                                   save_weights_only=True,
                                                   verbose=0) 
```

```python
# Using mixed precision training 
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy(policy="mixed_float16")
```

```python
# Building a model
from tensorflow.keras import layers

# Create base model 
input_shape=(224,224,3)
base_model=tf.keras.applications.EfficientNetB0(include_top=False)#Dont include the classification layers.
base_model.trainable=False# Freeze base model layers.


# Create the functional model
inputs=layers.Input(shape=input_shape, 
                    name="Input_layer")
# EfficientNet model has a rescaling built in 
# x = layers.Rescaling(1./255)(x)
x=base_model(inputs, 
             training=False)
x=layers.GlobalAveragePooling2D(name="pooling_layer")(x)
x=layers.Dense(len(class_names))(x)# having one neuron per class

#Seperate activation for output layers for converting into float32
outputs=layers.Activation("softmax",
                          dtype=tf.float32,
                          name="softmax_float32"
                         )(x)
model=tf.keras.Model(inputs,
                     outputs)# inputs and outputs are already defined but we need to bind togther

#Compile the model
model.compile(loss="sparse_categorical_crossentropy",
             optimizer=tf.keras.optimizers.Adam(),
             metrics=["accuracy"])
```

```python
# 1. Create a function to recreate the original model
def create_model():
  # Create base model
  input_shape = (224, 224, 3)
  base_model = tf.keras.applications.efficientnet.EfficientNetB0(include_top=False)
  base_model.trainable = False # freeze base model layers

  # Create Functional model 
  inputs = layers.Input(shape=input_shape, name="input_layer")
  # Note: EfficientNetBX models have rescaling built-in but if your model didn't you could have a layer like below
  # x = layers.Rescaling(1./255)(x)
  x = base_model(inputs, training=False) # set base_model to inference mode only
  x = layers.GlobalAveragePooling2D(name="pooling_layer")(x)
  x = layers.Dense(len(class_names))(x) # want one output neuron per class 
  # Separate activation of output layer so we can output float32 activations
  outputs = layers.Activation("softmax", dtype=tf.float32, name="softmax_float32")(x) 
  model = tf.keras.Model(inputs, outputs)
  
  return model

# 2. Create and compile a new version of the original model (new weights)
created_model = create_model()
created_model.compile(loss="sparse_categorical_crossentropy",
                      optimizer=tf.keras.optimizers.Adam(),
                      metrics=["accuracy"])

# 3. Load the saved weights
created_model.load_weights(checkpoint_path)

# 4. Evaluate the model with loaded weights
results_created_model_with_loaded_weights = created_model.evaluate(test_data)
```
